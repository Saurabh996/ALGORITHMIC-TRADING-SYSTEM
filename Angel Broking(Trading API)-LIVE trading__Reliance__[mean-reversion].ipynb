```python
from datetime import datetime
```

```python
arr = []
buy = []
sell = []
profit = []
n = 15
flag = [0]
```

```python
def data_feed(p):
    #print("{} @ {}".format(p, datetime.now().strftime("%H:%M:%S")))
    arr.append(p)
    #print("arr length: {}".format(len(arr)))
    if len(arr) > n:
        sma = sum(arr[-n:])/n
        print("price: {}, sma: {}, flag: {}".format(p, sma, flag[0]))
        if p > sma and flag[0] == 0:
            print("Buy @ {} @ {}".format(p, datetime.now().strftime("%H:%M:%S")))
            flag[0] = 1
            buy.append(p)
        if p < sma and flag[0] == 1:
            print("Sell @ {} @ {}".format(p, datetime.now().strftime("%H:%M:%S")))
            flag[0] = 0
            sell.append(p)
            profit.append(sell[-1]-buy[-1])
```
